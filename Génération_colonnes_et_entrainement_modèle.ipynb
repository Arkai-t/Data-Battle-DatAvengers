Generate litologie images

```python
nb_columns = 125
crop_size = [2700, 532]

png_dir = "/content/drive/Shareddrives/Data Battle/svg new pattern/"

type_sed = ['1_gravel', '2_clay', '4_silt siltstone', '6_silt siltstone', '8_shale', 
            '10_shalk', '11_dolomite', '12_anhydrite', '13_coal', '16_tuff', '18_tuff',
            '20_salt', 'chalk_shell', 'clay', 'gravel', 'limestone', 'marl', 'sand']

strats = []
name_index = 0
for j in range(nb_columns):
  i = 0
  column = np.zeros(crop_size)
  strat = []

  while i < column.shape[0]: 
    t = np.random.randint(0, len(type_sed))   # Select sed type
    img_array = cv2.imread(f'{png_dir}{type_sed[t]}.png', cv2.IMREAD_UNCHANGED)
    img_array = img_array[:, :, 2]

    # Copy patern at the y index twice (to fill the width)
    max_y = min(i+img_array.shape[0], crop_size[0])
    column[i : max_y, 0 : img_array.shape[1]] = img_array[:max_y-i, :]
    column[i : max_y, img_array.shape[1] : 2*img_array.shape[1]] = img_array[:max_y-i, :]


    #TODO
    #Add noise pattern (133 width) in /content/drive/Shareddrives/Data Battle/svg new pattern/noise/

    strat.append({'type': type_sed[t], 'start_y': i, 'end_y': (i+img_array.shape[0])})
    #print(strat)

    #Blank
    i += 1
    i += (max_y-i)

  #Add black bar left and write
  bar_width = 5
  column[:,0:bar_width] = np.full((column.shape[0], bar_width), 0)
  column[:,column.shape[1]-bar_width:column.shape[1]] = np.full((column.shape[0], bar_width), 0)

  #Noise https://gist.github.com/Prasad9/28f6a2df8e8d463c6ddd040f4f6a028a?permalink_comment_id=2917165#gistcomment-2917165
  noise =  np.random.normal(loc=0, scale=1, size=column.shape)
  img2 = column*2
  column = np.clip(np.where(img2 <= 1, (img2*(1 + noise*0.4)), (1-img2+1)*(1 + noise*0.4)*-1 + 2)/2, 0, 255)

  #Blur
  #column = cv2.GaussianBlur(column, (15, 15), 0)

  #Crop + conversion
  data = im.fromarray(column[:crop_size[0],:crop_size[1]])
  
  if data.mode != 'RGB':
    data = data.convert('RGB')
  data.save("/content/drive/Shareddrives/Data Battle/generated_column/"+str(name_index)+".png")

  strats.append({'name': f'{str(name_index)}.png', 'strats': strat})
  name_index += 1
```

Generate box infos csv

```python
#Re-organize strats info 
betterStrats = []
# For each images
for s in strats:
  last_pattern = None
  # For each layers (pattern images)
  for e in s['strats']:
    x_min = 0 #Stored in case of shift/rotation on bigger image
    y_min = e['start_y']
    x_max = crop_size[1]
    y_max = e['end_y'] if e['end_y'] < crop_size[0] else crop_size[0]

    # If 2 same type follow each other, merge box
    if e['type'] == last_pattern:
        betterStrats[-1][5] = y_max
    else:
      betterStrats.append([s['name'], e['type'], x_min, y_min, x_max, y_max])
      last_pattern = e['type']
    

# print("\nRESULTS")
# for s in betterStrats:
#   print(s[1])


df = pd.DataFrame(betterStrats, columns=['img', 'litho_type', 'xmin', 'ymin', 'xmax', 'ymax'])
df.to_csv("/content/drive/Shareddrives/Data Battle/generated_column/data.csv", index=False)
```

YOLOv8 PyTorch TXT format

```python
#Generate .txt for each.png
for img in df.img.unique():
  s = ""
  for index, row in df[df.img == img].iterrows():
    cls = type_sed.index(row.litho_type)
    width = row.xmax/crop_size[1] - row.xmin / crop_size[1]
    width = width if width < 1 else 1

    height = row.ymax/crop_size[0] - row.ymin/crop_size[0]
    height = height if height < 1 else 1
    
    x = row.xmin/crop_size[1] + width/2

    y = row.ymin/crop_size[0] + height/2

    s += f'{str(cls)} {x} {y} {width} {height}\n'

  with open(f'/content/drive/Shareddrives/Data Battle/generated_column/{img[:-4]}.txt', 'w') as f:
    f.write(s) 

#Generate .yaml
with open('/content/drive/Shareddrives/Data Battle/dataset.yaml', 'w') as f:
  f.writelines(['path: /content/drive/Shareddrives/Data Battle/\n',
                'train: generated_column/train/\n',
                'val: generated_column/val/\n',
                '\n',
                'names:\n'])
  for i in range(0, len(type_sed)):
    f.write(f' {i}: {type_sed[i]}\n')
```

Split train/test

```python
split_value = 100
dir = '/content/drive/Shareddrives/Data Battle/generated_column/'

#create train and val folder
if not os.path.isdir(f'{dir}train/'):
  os.mkdir(f'{dir}train/')
if not os.path.isdir(f'{dir}val/'):
  os.mkdir(f'{dir}val/')


for f in os.listdir(dir):
  #skip other files
  if (f[-4:] != '.txt' and f[-4:] != '.png'):
    continue
  
  #move to train
  if(int(f[:-4]) < split_value):
    os.rename(f'{dir}{f}', f'{dir}train/{f}')
  
  #move to val
  else:
    os.rename(f'{dir}{f}', f'{dir}val/{f}')
```
